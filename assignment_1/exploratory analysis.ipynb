```python
sns.histplot(df['target'], kde=True)
plt.xlabel('Price per Night')
plt.ylabel('Frequency')
plt.title('Distribution of Price per Night')
plt.show()
```

```python
sns.scatterplot(x='property_lon', y='property_lat', data=df)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Properties')
plt.show()
```

```python
sns.boxplot(x='property_type', y='target', data=df)
plt.xlabel('Property Type')
plt.ylabel('Price per Night')
plt.title('Boxplot of Property Types vs Price per Night')
plt.xticks(rotation=45)
plt.show()
```

```python
numerical_features = ['property_max_guests', 'property_bathrooms', 'property_bedrooms', 'property_beds',
                      'property_sqfeet', 'host_nr_listings', 'booking_min_nights', 'booking_max_nights',
                      'booking_availability_30', 'booking_availability_60', 'booking_availability_90',
                      'booking_availability_365', 'reviews_num', 'reviews_rating', 'reviews_acc',
                      'reviews_cleanliness', 'reviews_checkin', 'reviews_communication', 'reviews_location',
                      'reviews_value', 'reviews_per_month', 'target']
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Identify outliers in price per night using boxplot
sns.boxplot(x='target', data=df)
plt.xlabel('Price per Night')
plt.title('Boxplot of Price per Night')
plt.show()
```

```python
# Identify inconsistencies in property types
sns.countplot(x='property_type', data=df)
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.title('Count of Property Types')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify inconsistencies in room types
sns.countplot(x='property_room_type', data=df)
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Count of Room Types')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='property_bed_type', data=df)
plt.xlabel('Bed Type')
plt.ylabel('Count')
plt.title('Count of Bed Types')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify inconsistencies in cancellation policies
sns.countplot(x='booking_cancel_policy', data=df)
plt.xlabel('Cancellation Policy')
plt.ylabel('Count')
plt.title('Count of Cancellation Policies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify inconsistencies in host response time
sns.countplot(x='host_response_time', data=df)
plt.xlabel('Host Response Time')
plt.ylabel('Count')
plt.title('Count of Host Response Times')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify inconsistencies in host verification schemes
sns.countplot(x='host_verified', data=df)
plt.xlabel('Host Verification Schemes')
plt.ylabel('Count')
plt.title('Count of Host Verification Schemes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Generate a scatter plot of price per night against number of reviews
sns.scatterplot(x='target', y='reviews_num', data=df)
plt.xlabel('Price per Night')
plt.ylabel('Number of Reviews')
plt.title('Scatter Plot of Price per Night vs. Number of Reviews')
plt.show()
```

```python
# Generate a bar chart of the top 10 property types
top10_property_types = df['property_type'].value_counts().nlargest(10)
sns.barplot(x=top10_property_types.index, y=top10_property_types.values)
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.title('Top 10 Property Types')
plt.xticks(rotation=45)
plt.show()
```
